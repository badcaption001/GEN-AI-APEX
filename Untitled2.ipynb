```python
from transformers import pipeline
```

```python
import pandas as pd
```

```python
data = pd.read_html('https://en.wikipedia.org/wiki/National_Institutes_of_Technology_(India)',header=0)
```

```python
data[2]
```

```python
from transformers import AutoTokenizer, AutoModelForDocumentQuestionAnswering,Pipeline
```

```python
model="google/tapas-base-finetuned-wtq"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
data=data.astype(str)
```

```python
table_qa_pipeline = pipeline("table-question-answering", model=model, tokenizer=tokenizer)
```

```python
result = table_qa_pipeline(table=data, query="who is younger?")
```
